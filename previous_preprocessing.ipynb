```python
big_data = {
    'Application Train': application_train,
    'POS CASH Balance': pos_cash,
    'Bureau': bureau,
    'Bureau Balance': bureau_balance,
    'Credit Card Balance': credit_card,
    'Installments Payments': installments,
    'Previous Application': previous_application
}

for name, data in big_data.items():
    print(f'{name}: {data.shape}')
```

### Null values

```python
print('Number of NULL:\n')
for name, data in big_data.items():
    print(f'{name}: {data.isnull().sum().sum()}')
```

```python
samples = []
null_counts = []
null_percentages = []

for name, data in big_data.items():
    n_samples = len(data)
    n_nulls = data.isnull().sum().sum()
    null_pct = (n_nulls / (data.shape[0] * data.shape[1])) * 100
    
    samples.append(n_samples)
    null_counts.append(n_nulls)
    null_percentages.append(null_pct)

summary_data = pd.DataFrame({
    'Number of Samples': samples,
    'Number of Null Values': null_counts,
    'Percentage of Null Values': null_percentages
}, index=big_data.keys())

summary_data['Percentage of Null Values'] = summary_data['Percentage of Null Values'].round(2)

print(summary_data)
```

### Missing

```python
def missing_values(data):
    missing_values_count = data.isnull().sum()
    missing_values_percentage = 100 * data.isnull().sum() / len(data)
    missing_values_df = pd.DataFrame({
        'Column Name': data.columns,
        'Missing Values': missing_values_count,
        'Percentage': missing_values_percentage
    }).reset_index(drop=True)
    
    return missing_values_df

train_missing = missing_values(application_train).sort_values(by='Percentage', ascending=False)
pos_cash_missing = missing_values(pos_cash).sort_values(by='Percentage', ascending=False)
bureau_missing = missing_values(bureau).sort_values(by='Percentage', ascending=False)
bureau_balance_missing = missing_values(bureau_balance).sort_values(by='Percentage', ascending=False)
credit_card_missing = missing_values(credit_card).sort_values(by='Percentage', ascending=False)
installments_missing = missing_values(installments).sort_values(by='Percentage', ascending=False)
previous_application_missing = missing_values(previous_application).sort_values(by='Percentage', ascending=False)
```
